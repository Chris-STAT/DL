```python
from keras.preprocessing.image import ImageDataGenerator
```

** Image generator **

```python
train_datagen = ImageDataGenerator(rescale= 1./255)
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
train_generator = train_datagen.flow_from_directory(
                     train_dir,
                     target_size = (150,150),
                     batch_size = 20, 
                     class_mode = 'categorical')
```

```python
validation_generator = test_datagen.flow_from_directory(
                       validation_dir,
                       target_size = (150, 150),
                       batch_size = 20,
                       class_mode = 'categorical')
```

```python
import keras
import numpy as np
from keras import Model
from keras.applications import vgg16
from keras.models import Sequential
from keras.layers import Activation, Flatten, Dense, Dropout, Conv2D, MaxPooling2D
```

** Import pre-trained VGG model (transfer learning) **

```python
input_shape = (150, 150,3)
vgg_model = vgg16.VGG16(include_top = False, weights = 'imagenet', input_shape = input_shape)
```

** Freeze VGG layers (set VGG layers as non-trainable) **

```python
vgg_model.trainable = False 
for layer in vgg_model.layers:
    layer.trainable = False
```

```python
vgg_model.summary()
```

```python
#last_layer = vgg_model.output
#new_layer = Flatten()(last_layer)
#new_layer = Dense(256, activation = 'relu')(new_layer)
#new_layer = Dropout(0.5)(new_layer)
#preds = Dense(6, activation = 'sigmoid')(new_layer)
#model = Model(vgg_model.input, preds)
#model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

** Add news layers and train the model on the geological datasets **

```python
last_layer = vgg_model.output
new_layer = Conv2D(16,(4,4), padding = 'same', activation = 'relu')(last_layer)
new_layer = MaxPooling2D((2,2))(new_layer)
new_layer = Flatten()(new_layer)
new_layer = Dense(256, activation = 'relu')(new_layer)
new_layer = Dropout(0.5)(new_layer)
preds = Dense(6, activation = 'sigmoid')(new_layer)
model = Model(vgg_model.input, preds)
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

** Fit the model **

```python
history = model.fit_generator(
         train_generator,
         steps_per_epoch = 100,
         epochs = 30,
         validation_data = validation_generator,
         validation_steps =50)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

** Remove the model output layer and output the vector representation of each input image **

```python
model.layers.pop()
model_sim = Model(model.input, model.layers[-1].output)
```

```python
def img_gen(orig_img):
      new_img = image.load_img(orig_img, target_size = (150,150))
      new_img = image.img_to_array(new_img).astype('float32')
      new_img  = new_img/255.  
      new_img = new_img.reshape((1,150,150,3))
      return(new_img)
```

```python
from scipy import spatial
```

** Calculate cosine similarity **

```python
def img_sim(img_vec_target, img_vec_other):
       return(1-spatial.distance.cosine(img_vec_target, img_vec_other))
```

** Return the indexes of the top-K similar images **

```python
def top_N_sim_imgs(k, target_img, candidate_imgs, model):
        
        similarity = []
        
        target_value = model.predict(target_img)
        
        for i in range(len(candidate_imgs)):
            
            candidate_value = model.predict(candidate_imgs[i])
            similarity.append(img_sim(target_value, candidate_value))
         
        top_k = sorted(sorted(range(len(similarity)), key = lambda i: similarity[i])[-k:])
        return(top_k)
```
